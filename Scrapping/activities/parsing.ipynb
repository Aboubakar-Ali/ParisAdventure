```python
import pandas as pd
import re


df = pd.read_csv('activities.csv')

#Price
# clean price
def clean_price(price):
    if "à partir de" in str(price):
        return price.replace("à partir de ", "")
    else:
        return price


df = df.rename(columns={'prix': 'price'})
df['price'] = df['price'].apply(clean_price)
df['price'] = df['price'].replace('Prix non disponible', 0)
df['price'] = df['price'].str.replace('€', '').str.strip()
df['price'] = df['price'].str.replace(',', '.').astype(float)


# Open 
df = df.rename(columns={'open': 'open'})
df['open'] = df['open'].replace('Ouverture non disponible', '09:00 - 21:00')
df['open'] = df['open'].str.replace(' à ', ' - ')
df['open'] = df['open'].str.replace('^De ', '', regex=True)

# open and close
df[['Open', 'Close']] = df['open'].str.split(' - ', expand=True)
df['Open'] = df['Open'].str.replace(':', '.').astype(float)
df['Close'] = df['Close'].str.replace(':', '.').astype(float)

df = df.drop(columns=['open'])






# #time
# Supprimer les lignes avec de longues chaînes 
df = df[df['time'].str.len() <= 50]

def extract_max_time(time_str):
    # Extraire tous les nombres de la chaîne
    numbers = [float(s.replace(',', '.')) for s in re.findall(r'\d+,\d+|\d+', time_str)]
    
    # Prendre le maximum
    if numbers:
        return max(numbers)
    else:
        return None

df['time'] = df['time'].apply(extract_max_time)


df['time'] = df['time'].replace("Moins d'une heure", 1.0)

df = df.rename(columns={'time': 'time'})
df = df.rename(columns={'note': 'note'})
df = df.rename(columns={'Nom': 'name'})
df = df.rename(columns={'adress': 'address'})
df['price'] = df['price'].fillna(0.0)
df['time'] = df['time'].fillna(0.0)


df.to_csv('Clean.csv', index=False)

df




```
