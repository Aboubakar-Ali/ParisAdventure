```python
import pandas as pd

# get data from csv file
df = pd.read_csv('restaurants.csv')

df = df[df['FOURCHETTE DE PRIX'].str.len() <= 9]
df['FOURCHETTE DE PRIX'] = df['FOURCHETTE DE PRIX'].replace("Française", "0€ - 0€")

#  min et max
df['prix_min'] = df['FOURCHETTE DE PRIX'].str.extract(r'(\d+\.?\d*)€')[0].str.replace(',', '.').astype(float)
df['prix_max'] = df['FOURCHETTE DE PRIX'].str.extract(r'- (\d+\.?\d*)€')[0].str.replace(',', '.').astype(float)


df = df.drop(columns=['FOURCHETTE DE PRIX'])

df = df.rename(columns={'ADRESSE': 'address'})
df = df.rename(columns={'NOTE': 'note'})
df = df.rename(columns={'prix_min': 'price_min'})
df = df.rename(columns={'prix_max': 'price_max'})
df = df.rename(columns={'NOM': 'name'})

# make note column float
df['note'] = df['note'].str.extract(r'(\d+\.?\d*)')[0].str.replace(',', '.').astype(float)

df.to_csv('Clean.csv', index=False)
df
```
